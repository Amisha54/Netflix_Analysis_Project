How many movies and shows are in the dataset? Show the percentage and  Top 10 Popular Geners on Netflix 

```python
data['ContentType'].value_counts()
```

```python
sns.countplot(data=data,x='ContentType')
fig = plt.gcf()
fig.set_size_inches(7, 5)
plt.title('Netflix')
plt.xlabel("Netflix (TVSeries/Movie)")
plt.ylabel("Total Count")
plt.show()
```

```python
total_count = len(data)
content_type_counts = data['ContentType'].value_counts()
percentage = (content_type_counts / total_count) * 100

plt.pie(percentage, explode=(0.025, 0.025), labels=percentage.index, colors=['yellow', 'orange'],
        autopct='%1.1f%%', startangle=180)
plt.title('Percentage of Netflix Content Types')
plt.show()
```

```python
sns.countplot(data=data, y='Genre', order=data['Genre'].value_counts().head(10).index)
plt.title('Top 10 Genres on Netflix')
plt.show()
```
